# Micro-Farm

## Introducció

Cal desenvolupar una granja virtual, segons aquestes especificacions.

A totes les pantalles hi haurà 4 parts:

- La part superior mostra el nom de la granja, i els diners / nivell
```
Exemple: Micro-Farm              200€ / 01
```
- Element principal de la pantalla (menu, camps, etc ...)
- Missatges de 'prompt' amb la informació de les últimes accions (4 línies)
- Entrada de peticions de l'usuari (1 línia)

Quan s'escull una opció incorrecte s'afegeix el missatge al prompt:
La opció 'xxxx' no és vàlida

Endavant es suposa que el nom de la granja és 'Micro-Farm'

**IMPORTANT**: Si a algun dels exmples, no es mostra el prompt entre l'element principal i l'entrada de peticions, igualment s'hi ha de posar

## Menú principal:

Hi haurà un menú principial amb les següents opcions:

```
Micro-Farm              200€ / 01
1 Nom de la granja
2 Plantacions             40 / 10
3 Animals                 05 / 02
4 Fabriques               01 / 10
5 Graner                  20 / 25
6 Estadístiques
0 Sortir
Prompt 0
Prompt 1
Prompt 2
Prompt 3
Escull una opció [0, 6]:
```

Si el nom de la granja encara no està definit, la primera opció mostra:

```
????                     000 / 00
1 Nom de la granja   (no definit)
i és la única que es pot escollir (a part de la 0)
```

Al menú principal, els valors numèrics representen XX / YY:

- Les primeres opcions XX són el número d'elements que ja estàn llestos
- Les segones opcions YY són el número d'elements que es poden elaborar/demanar

A l'exemple, hi ha 40 plantacions per recollir i 10 plantacions disponibles
A l'exemple, hi ha 5 productes d'animals per recollir i 2 animals per alimentar
A l'exemple, el graner té 20 elements de 50 màxims

## Nom de la granja:

El nom de la granja permet lletres, espais i números però no caràcters especials

## Plantacions:

La opció plantacions mostrarà el camp amb les plantacions, 
hi ha 50 terrenys disponibles que s'han de mostrar en una graella de 10x5

```
Micro-Farm              200€ / 01
Plantacions:              40 / 10
BBBBBXXP33
2233000000
0000000000
0000000000
0000000000
Prompt 0
Prompt 1
Prompt 2
Prompt 3
Escriu quina acció vols realitzar (plantar, recollir, sortir):
```

Al camp:
- 0 és un espai disponible per plantar
- B és blat llest per recollir
- X és panotxa llesta per recollir
- P és patata llesta per recollir
- Els números és el nombre de torns que cal esperar per recollir aquell element

Per plantar l'usuari haurà d'escriure: plantar blat 45
Aleshores s'afegeix el missatge al prompt: S'han plantat 45 blats
Per recollir l'usuari haurà d'escriure: recollir blat
Aleshores s'afegeix el missatge al prompt: S'han recollit 45 blats
Si no hi ha terrenys disponibles s'ecriu: No hi ha terrenys disponibles

Els tipus de plantacions són: blat, panotxa, patata

Les llavors són infinites, es pot plantar sempre que hi hagi espai al camp
Un cop plantades:
- El blat triga 5 interaccions a créixer
- La panotxa triga 8 interaccions a créixer
- La patata triga 9 interaccions a créixer

Quan es recull una plantació, aquesta es guarda al graner
Quan es recull una plantació es suma 1 Euro per cada 'terreny'

## Animals:
```
Micro-Farm              200€ / 01
Animals:                  40 / 10
V0 V1 G0 G1 G2 X0 X1
M9 L3 M5 O2 O3 N1 M1
Prompt 0
Prompt 1
Prompt 2
Prompt 3
Escriu quina acció vols realitzar (alimentar, recollir, sortir):
```
La opció animals mostrarà el camp amb els animals, inicialment hi ha:

- 2 vaques (V0 i V1)
- 3 gallines (G0, G1 i G2)
- 2 xais (X0 i X1)

Els valors que mostra cada animal són:

-M9: Vol dir que l'animal està menjant i que falten 9 interaccions per acabar
-L3: Vol dir que hi a tres unitats de llet llestes per recollir
-O2: Vol dir que hi a dues unitats d'ous llestos per recollir
-N1: Vol dir que hi a una unitat de llana llesta per recollir

Les vaques s'alimenten amb 3 unitats de panotxa
Les gallines s'alimenten amb 5 unitats de blat
Els xaix s'alimenten amb 2 unitats de patata

Per alimentar l'usuari haurà d'escriure: alimentar V0
Aleshores s'afegeix el missatge al prompt: S'ha alimentat V0
Per recollir l'usuari haurà d'escriure: recollir V0
Aleshores s'afegeix el missatge al prompt: S'ha recollit V0
Si no hi ha prou menjar per alimentar s'ecriu: No hi ha prou menjar per alimentar V0
Si no hi ha res que recollir s'ecriu: No hi ha res per recollir de V0
Si el graner ja sobrepassa el limit s'ecriu: No es pot recollir V0, el graner està ple
Si no hi ha panotxa per alimentar les vaques s'ecriu: No hi ha panotxa per alimentar V0
Si no hi ha blat per alimentar les gallines s'ecriu: No hi ha blat per alimentar G0
Si no hi ha patata per alimentar els xais s'ecriu: No hi ha patata per alimentar X0

Es sumen 2 Euros per cada 'litre' de llet recollit
Es suma 1 Euro per cada ou recollit
Es sumen 3 Euros per cada unitat de llana recollida

## Fàbriques

Les fabriques s'han de comprar amb diners, originalment mostrarà:
```
Micro-Farm              200€ / 01
Fàbriques:                 2 / 10
Papas fregides      (comprar 25€)
Ganxets             (comprar 50€)
Formatge           (comprar  75€)
Iogurt             (comprar 100€)
Truites            (comprar 125€)
Batuts             (comprar 150€)
Llana              (comprar 175€)
Roba               (comprar 200€)
Prompt 0
Prompt 1
Prompt 2
Prompt 3
Escriu quina acció vols realitzar (comprar, fabricar, sortir):
```
Quan es compra alguna fàbrica es mostra:
```
Micro-Farm              200€ / 01
Fàbriques:                 2 / 10
Papas fregides       Fabricant 15
Ganxets             (comprar 50€)
Formatge                    Llest
Iogurt               Fabricant 25
Truites            (comprar 125€)
Batuts             (comprar 150€)
Cabdell            (comprar 150€)
Roba               (comprar 200€)
Prompt 0
Prompt 1
Prompt 2
Prompt 3
Escriu quina acció vols realitzar (comprar, fabricar, valors, sortir):
```
El numero de fabricació és el número de torns que cal esperar per tenir el producte acabat
Quan s'ha acabat de fabricar es mostra llest

Per comprar una fàbrica l'usuari haurà d'escriure: comprar 'Papas fregides'
Aleshores s'afegeix el missatge al prompt: S'ha comprat 'Papas fregides'
Si no hi ha diners suficients s'ecriu: No hi ha diners suficients per comprar 'Papas fregides'
Si ja s'ha comprat s'ecriu: Ja s'ha comprat 'Papas fregides'

Si es vol fabricar un producte l'usuari haurà d'escriure: fabricar 'Papas fregides'
Aleshores s'afegeix el missatge al prompt: S'ha començat a fabricar 'Papas fregides'
Si no hi ha prou producte per fabricar s'ecriu: No hi ha prou producte per fabricar 'Papas fregides'
Si ja s'està fabricant s'ecriu: Ja s'està fabricant 'Papas fregides'

Si es vol recollir un producte l'usuari haurà d'escriure: recollir 'Papas fregides'
Aleshores s'afegeix el missatge al prompt: S'ha recollit 'Papas fregides'
Si no hi ha res per recollir s'ecriu: No hi ha res per recollir de 'Papas fregides'
Si el graner ja sobrepassa el limit s'ecriu: No es pot recollir 'Papas fregides', el graner està ple

Originalment, per fabricar al:
- Papas fregides cal 5 unitats de patata
- Ganxets cal 2 unitats de patates i 1 unitat de panotxa
- Formatge cal 3 unitats de llet
- Iogurt cal 2 unitats de llet
- Truites cal 5 unitats d'ous i 2 unitats de patata
- Batuts cal 2 unitats de llet i 1 unitat d'ous
- Cabdell cal 2 unitats de llana
- Roba cal 3 unitats de llana i 1 unitat de panotxa

Originalment, els temps de fabricació són, en interaccions:
- Papas fregides 25
- Ganxets 15
- Formatge 30
- Iogurt 10
- Truites 15
- Batuts 20
- Cabdell 15
- Roba 25

Originalment, quan es recull un producte es suma:
- Papas fregides 2€
- Ganxets 1€
- Formatge 2€
- Iogurt 1€
- Truites 5€
- Batuts 4€
- Cabdell 10€
- Roba 20€

Valors mostra la següent taula:
```
Micro-Farm              200€ / 01
Valors Fàbriques:
Producte         Temps      Valor
Papas fregides      25         2€
Ganxets             15         1€
Formatge            30         2€
Iogurt              10         1€
Truites             15         5€
Batuts              20         4€
Cabdell             15        10€
Roba                25        20€
Prompt 0
Prompt 1
Prompt 2
Prompt 3
Escull una opció per tornar (fabriques, menu):
```
**IMPORTANT**: A cada canvi de nivell, els temps s'augmenten 5 i els valors 1
