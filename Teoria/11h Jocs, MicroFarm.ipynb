# MicroFarm 

## Introducció

Cal desenvolupar una granja virtual, segons aquestes especificacions.

A totes les pantalles hi haurà 4 parts:

- La part superior mostra el nom de la granja, i els diners / nivell
- Element principal de la pantalla (menu, camps, etc ...)
- Missatges de 'prompt' amb la informació de les últimes accions (4 línies)
- Entrada de peticions de l'usuari (1 línia)

Quan s'escull una opció incorrecte s'afegeix el missatge al prompt:
La opció 'xxxx' no és vàlida

A cada interacció s'esborra la pantalla i es torna a dibuixar tot (part superior, element principal, prompt i entrada)

Endavant es suposa que el nom de la granja és 'MicroFarm '

**IMPORTANT**: Si a algun dels exemples, no es mostra el prompt entre l'element principal i l'entrada de peticions, igualment s'hi ha de posar

## Menú principal:

Hi haurà un menú principial amb les següents opcions:

```text
MicroFarm               200€ / 01
1 Nom de la granja
2 Plantacions             40 / 10
3 Animals                 03 / 02
4 Fabriques               01 / 10
5 Graner                  20 / 99
6 Estadístiques
7 Trampa
0 Sortir
Prompt 0
Prompt 1
Prompt 2
Prompt 3
Escull una opció [0, 7]:
```

Si el nom de la granja encara no està definit, la primera opció mostra:

```text
????                     000 / 00
1 Nom de la granja   (no definit)
```

En aquest cas (no s'ha definit el nom), només es poden escollir les opcions 1 i 0.

Al menú principal, els valors numèrics representen XX / YY:

- Les primeres opcions XX són el número d'elements que ja estàn llestos

- Les segones opcions YY són el número d'elements que es poden elaborar/demanar

    A l'exemple, hi ha 40 plantacions per recollir i 10 plantacions disponibles

    A l'exemple, hi ha 3 productes d'animals per recollir i 2 animals per alimentar

    A l'exemple, el graner té 20 elements de 99 màxims

## Nom de la granja:

El nom de la granja permet lletres, espais i números però no caràcters especials

## Plantacions:

La opció plantacions mostrarà el camp amb les plantacions, hi ha 50 terrenys disponibles que s'han de mostrar en una graella de 10x5

```text
MicroFarm               200€ / 01
Plantacions:              40 / 10
BBBBBXXP33
2233000000
0000000000
0000000000
0000000000
Prompt 0
Prompt 1
Prompt 2
Prompt 3
Escriu quina acció vols realitzar (ajuda, menu):
```

#### Explicació de les plantacions:

- 0 és un espai disponible per plantar
- B és blat llest per recollir
- X és panotxa llesta per recollir
- P és patata llesta per recollir
- Els números és el nombre de torns que cal esperar per recollir aquell element


#### Accions de les plantacions:

- Per plantar l'usuari haurà d'escriure: plantar blat 45

    Aleshores s'afegeix el missatge al prompt: S'han plantat 45 blats

- Per recollir l'usuari haurà d'escriure: recollir blat

    Aleshores s'afegeix el missatge al prompt: S'han recollit 45 blats

    Si no hi ha terrenys disponibles s'ecriu: No hi ha terrenys disponibles

Els tipus de plantacions són: blat, panotxa, patata

Les llavors són infinites, es pot plantar sempre que hi hagi espai al camp, un cop plantades:

#### Valors de les plantacions:

- El blat triga 5 interaccions a créixer
- La panotxa triga 8 interaccions a créixer
- La patata triga 9 interaccions a créixer

Quan es recull una plantació, aquesta es guarda al graner

#### Diners recollits amb les plantacions:

Quan es recull una plantació es suma 1 Euro per cada 'terreny'

#### Canvi de nivell, comprar terrenys:

A cada planta de nivell, es poden comprar 10 terrenys més fins a un màxim de 100

- Cada terreny costarà 10€

- Per comprar terrenys l'usuari haurà d'escriure: comprar terrenys

    Aleshores s'afegeix el missatge al prompt: S'han comprat 10 terrenys per valor de 100€

    Si no hi ha diners suficients s'ecriu: Fan falta 100€ per comprar terrenys

    Si no es poden comprar més terrenys per aquest nivell s'ecriu: El màxim en aquest nivell és de 70 terrenys.

    Si no es poden comprar terrenys s'ecriu: Ja tens 100 terrenys, no se'n poden comprar més

#### Opció 'ajuda' a les plantacions:

L'opció 'ajuda' mostra exemples de peticions:

```text
plantar blat 45 (blat, panotxa, patata)
recollir patata (blat, panotxa, patata)
comprar terrenys
```

## Animals:
```text
MicroFarm               200€ / 01
Animals:                  40 / 10
V0 Vaca, menjant 9
V1 Vaca, 3 unitats de llet
G0 Gallina, menjant 5
G1 Gallina, 2 unitats d'ous
G2 Gallina, esperant
X0 Xai, menjant 7
X1 Xai, 1 unitat de llana
Prompt 0
Prompt 1
Prompt 2
Prompt 3
Escriu quina acció vols realitzar (ajuda, menu):
```

#### Explicació de l'apartat d'animals:

La opció animals mostrarà el camp amb els animals, inicialment hi ha:

- 2 vaques (V0 i V1)
- 3 gallines (G0, G1 i G2)
- 2 xais (X0 i X1)

Els valors que mostra cada animal són:

- menjant X: Vol dir que l'animal està menjant i que falten X interaccions per acabar
- 3 unitats de llet: Vol dir que hi a tres unitats de llet llestes per recollir
- 2 unitats d'ous: Vol dir que hi a dues unitats d'ous llestos per recollir
- 1 unitat de llana: Vol dir que hi a una unitat de llana llesta per recollir
- esperant: Vol dir que l'animal no està menjant i no té cap unitat per recollir

#### Accions a l'apartat d'animals:

- Es pot demanar el cost d'alimentar els animals amb: pinso vaques

    Aleshores s'afegeix el missatge al prompt: El cost d'alimentar les vaques és de 3 unitats de panotxa

    Per les gallines: El cost d'alimentar les gallines és de 5 unitats de blat

    Per els xais: El cost d'alimentar els xais és de 2 unitats de patata

- Per alimentar l'usuari haurà d'escriure: alimentar V0

    Aleshores s'afegeix el missatge al prompt: S'ha alimentat V0

    Si no hi ha prou menjar per alimentar s'ecriu: No hi ha prou menjar per alimentar V0

    Si no hi ha panotxa per alimentar les vaques s'ecriu: No hi ha panotxa per alimentar V0

    Si no hi ha blat per alimentar les gallines s'ecriu: No hi ha blat per alimentar G0

    Si no hi ha patata per alimentar els xais s'ecriu: No hi ha patata per alimentar X0

- Per recollir l'usuari haurà d'escriure: recollir V0

    Aleshores s'afegeix el missatge al prompt: S'ha recollit V0

    Si no hi ha res que recollir s'ecriu: No hi ha res per recollir de V0

- Si el graner ja sobrepassa el limit s'ecriu: No es pot recollir V0, el graner està ple

#### Diners recollits a l'apartat d'animals:

- Es sumen 2 Euros per cada 'litre' de llet recollit
- Es suma 1 Euro per cada ou recollit
- Es sumen 3 Euros per cada unitat de llana recollida

#### Opció 'ajuda' a l'apartat d'animals:

L'opció 'ajuda' mostra exemples de peticions:

```text
pinso vaques (vaques, gallines, xais)
alimentar V0
recollir G1
```
## Fàbriques

Les fabriques s'han de comprar amb diners, originalment mostrarà:
```text
MicroFarm               200€ / 01
Fàbriques:                 2 / 10
Patates fregides    (comprar 25€)
Ganxets             (comprar 50€)
Formatge           (comprar  75€)
Iogurt             (comprar 100€)
Truites            (comprar 125€)
Batuts             (comprar 150€)
Llana              (comprar 175€)
Roba               (comprar 200€)
Prompt 0
Prompt 1
Prompt 2
Prompt 3
Escriu quina acció vols realitzar (ajuda, menu):
```
Quan es compra alguna fàbrica es mostra:
```text
MicroFarm               200€ / 01
Fàbriques:                 2 / 10
Patates fregides     Fabricant 15
Ganxets             (comprar 50€)
Formatge                    Llest
Iogurt               Fabricant 25
Truites            (comprar 125€)
Batuts             (comprar 150€)
Cabdell            (comprar 150€)
Roba               (comprar 200€)
Prompt 0
Prompt 1
Prompt 2
Prompt 3
Escriu quina acció vols realitzar (ajuda, menu):
```

#### Accions de les fàbriques:

- El numero de fabricació és el número de torns que cal esperar per tenir el producte acabat

    Quan s'ha acabat de fabricar es mostra llest

- Per comprar una fàbrica l'usuari haurà d'escriure: **comprar fregides**

    Aleshores s'afegeix el missatge al prompt: S'ha comprat 'Patates fregides'

    Si no hi ha diners suficients s'ecriu: No hi ha diners suficients per comprar 'Patates fregides'

    Si ja s'ha comprat s'ecriu: Ja s'ha comprat 'Patates fregides'

- Si es vol fabricar un producte l'usuari haurà d'escriure: **fabricar Patates fregides**

    Aleshores s'afegeix el missatge al prompt: S'ha començat a fabricar 'Patates fregides'

    Si no hi ha prou producte per fabricar s'ecriu: No hi ha prou producte per fabricar 'Patates fregides'

    Si ja s'està fabricant s'ecriu: Ja s'està fabricant 'Patates fregides'

- Si es vol recollir un producte l'usuari haurà d'escriure: **recollir Patates fregides**

    Aleshores s'afegeix el missatge al prompt: S'ha recollit 'Patates fregides'
    
    Si no hi ha res per recollir s'ecriu: No hi ha res per recollir de 'Patates fregides'
    
    Si el graner ja sobrepassa el limit s'ecriu: No es pot recollir 'Patates fregides', el graner està ple

## Estadístiques:
```text
MicroFarm               200€ / 01
Camps recollits              2000
Aliments produïts            1000
Productes fabricats           500
Productes venuts              100
Nivell                          2
Prompt 0
Prompt 1
Prompt 2
Prompt 3
Escull una opció per tornar (reiniciar, menu):
```
La opció 'reiniciar' torna a començar el joc i esborra tot el progrés

## Trampa:

La opció trampa permet canviar el nivell del joc, per exemple:

```text
MicroFarm               200€ / 01
Nivell actual: 2
Prompt 0
Prompt 1
Prompt 2
Prompt 3
Escriu el nivell al que vols canviar ([0, 9], graner, menu):
```

- Amb un valor numèric entre 0 i 9, automàticament hi haurà els diners necessaris per provocar el canvi de nivell. I s'escriu al prompt:

    Has fet trampa fins al nivell X (on X és el número de nivell)

- Amb la opció 'graner' es buidarà el graner a la meitat. I s'escriu al prompt:

    Has fet trampa, el graner s'ha buidat a la meitat

**IMPORTANT**: Si es fa trampa per més d'un nivell, per exemple es passa del 2 al 5, 
s'han d'incrementar els temps de fabricació i els valors de les fabriques de manera acumulativa com 
si s'hagués passat per tots els nivells intermitjos.

## Canvis de nivell:

Els canvis de nivell es produeixen a l'arribar per primera vegada a un valor concret de diners, 
encara que més tard es gastin diners, es manté el nivell.

- Originalment nivell 0
- Primer nivell 10€
- Segon nivell 50€
- Tercer nivell 100€
- Quart nivell 200€
- Cinquè nivell 500€
- Sisè nivell 1000€
- Setè nivell 2000€
- Vuitè nivell 5000€
- Novè nivell 10000€

Quan hi ha un canvi de nivell s'escriu al prompt: Has pujat al nivell X (on X és el número de nivell)

## Recomanacions:

- Separeu la lògica de funcionament del mètodes de dibuix
- Organitzeu els diccionaris de dades que necessiteu
- Feu almenys una funció per cada opció del menú
- Teniu en compte que cada interacció canvia l'estat del joc (i dels diccionaris)
- Teniu en compte que cada canvi de nivell canvia els temps i els valors de les fàbriques
- Practiqueu per l'exàmen, feu servir: diccionaris, llistes, funcions, bucles, condicionals, ...
